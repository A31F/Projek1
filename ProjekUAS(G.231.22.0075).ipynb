```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
#mengupload file
uploaded = files.upload()
```

```python
#membaca folder csv
data = pd.read_csv("stock1 (1).csv")

#memberi nomer urut
data = data.set_index(pd.Index(range(1, len(data)+1)))

#menampilkan
data
```

```python
plt.figure(figsize=(10, 5))

#memilih data Purchase Price
plt.hist(data['Purchase Price'], bins=13, edgecolor='white')

#memberi keterangan nama / judul
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Price')

#menampilkan
plt.show()
```

```python
plt.figure(figsize=(10, 5))

#memilih data Current Price
plt.hist(data['Current Price'], bins=13, edgecolor='white')

#memberi nama keterangan / judul
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.title('Histogram of Current Price')

#menampilkan
plt.show()
```

```python
plt.figure(figsize=(14, 5))

#memilih data Gain/Loss
plt.hist(data['Gain/Loss'], bins=13, edgecolor='white')

#memberi keterangan / judul
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Histogram of Gain/Loss')

#menampilkan
plt.show()
```

```python
#Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi

#memilih data Shares tertinggi
sorted_data = data.sort_values('Shares', ascending=False)
data5 = sorted_data.head(5)

#memberi nomer urut
data5 = data5.set_index(pd.Index(range(1, len(data5)+1)))

#memberi nama tabel
result = data5[['Stock Name', 'Symbol','Shares']]

#menampilkan hasil
print(result)
```
